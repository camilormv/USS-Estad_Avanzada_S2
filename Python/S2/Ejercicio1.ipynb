Ejemplo: Supongamos que tenemos una población de estudiantes y queremos estudiar su altura promedio. Seleccionamos 30 muestras aleatorias de 50 estudiantes cada una y calculamos la media de cada muestra. Según el Teorema del Límite Central, si repetimos este proceso un gran número de veces, la distribución de las medias muestrales se aproximará a una distribución normal, independientemente de la forma de la distribución original de las alturas.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Generar una población de 10,000 estudiantes con alturas entre 140 cm y 200 cm
poblacion = np.random.uniform(140, 200, 10000)

# Número de muestras y tamaño de cada muestra
num_muestras = 30
tamaño_muestra = 50

# Almacenar las medias muestrales
medias_muestrales = np.array([np.mean(np.random.choice(poblacion, tamaño_muestra)) for _ in range(num_muestras)])

# Configuración de la gráfica
plt.figure(figsize=(10, 6))
sns.histplot(medias_muestrales, kde=True, bins=20, color='blue', stat='density')

# Superponer una distribución normal
media_poblacional = np.mean(poblacion)
std_error = np.std(poblacion) / np.sqrt(tamaño_muestra)
x = np.linspace(media_poblacional - 4*std_error, media_poblacional + 4*std_error, 100)
plt.plot(x, stats.norm.pdf(x, media_poblacional, std_error), 'r--', lw=2)

plt.title('Distribución de las Medias Muestrales')
plt.xlabel('Medias Muestrales')
plt.ylabel('Densidad')
plt.legend(['Distribución Normal', 'Medias Muestrales'])
plt.show()
```

Ejercicio 2: Ejemplo: Queremos estimar el ingreso promedio de los hogares en una ciudad. Utilizamos el método de estimación por intervalos. Seleccionamos una muestra aleatoria de 100 hogares y calculamos el ingreso promedio y
la desviación estándar. Con estos datos, podemos construir un intervalo de confianza para el ingreso promedio de toda la ciudad. 
Seleccionamos una muestra aleatoria de 100 hogares y calculamos el ingreso promedio y la desviación estándar. Con estos datos, podemos construir un intervalo de confianza para el ingreso promedio de toda la ciudad.

```python
import numpy as np
from scipy import stats

# Generar una población de 10,000 hogares con ingresos distribuidos normalmente
np.random.seed(0)  # para reproducibilidad
poblacion_ingresos = np.random.normal(50000, 15000, 10000)

# Seleccionar una muestra aleatoria de 100 hogares
tamaño_muestra = 100
muestra_ingresos = np.random.choice(poblacion_ingresos, tamaño_muestra)

# Calcular el ingreso promedio y la desviación estándar de la muestra
ingreso_promedio_muestra = np.mean(muestra_ingresos)
desviación_estándar_muestra = np.std(muestra_ingresos, ddof=1)

# Nivel de confianza del 95%
nivel_confianza = 0.95
alfa = 1 - nivel_confianza

# Calcular el valor crítico z
z_critical = stats.norm.ppf(1 - alfa/2)

# Calcular el error estándar
error_estándar = desviación_estándar_muestra / np.sqrt(tamaño_muestra)

# Calcular el intervalo de confianza
margen_error = z_critical * error_estándar
intervalo_confianza = (ingreso_promedio_muestra - margen_error, ingreso_promedio_muestra + margen_error)

# Mostrar los resultados
print(f"Ingreso promedio de la muestra: ${ingreso_promedio_muestra:.2f}")
print(f"Desviación estándar de la muestra: ${desviación_estándar_muestra:.2f}")
print(f"Intervalo de confianza del 95% para el ingreso promedio: ${intervalo_confianza[0]:.2f} a ${intervalo_confianza[1]:.2f}")
```

```python
import numpy as np
from scipy import stats

# Datos de la muestra
media_muestral = 120  # mmHg
desviacion_estandar_muestral = 15  # mmHg
tamaño_muestra = 200

# Nivel de confianza del 95%
nivel_confianza = 0.95
alfa = 1 - nivel_confianza
z_critical = stats.norm.ppf(1 - alfa/2)

# Error estándar
error_estándar = desviacion_estandar_muestral / np.sqrt(tamaño_muestra)

# Intervalo de confianza
margen_error = z_critical * error_estándar
intervalo_confianza = (media_muestral - margen_error, media_muestral + margen_error)

# Resultados
print(f"Media muestral de la presión arterial: {media_muestral} mmHg")
print(f"Desviación estándar muestral: {desviacion_estandar_muestral} mmHg")
print(f"Intervalo de confianza del 95% para la presión arterial promedio: {intervalo_confianza[0]:.2f} mmHg a {intervalo_confianza[1]:.2f} mmHg")
```

El problema consiste en analizar un conjunto de datos de ventas y calcular algunas métricas básicas, como:

El total de ventas.
El promedio de ventas por producto.
Generar un gráfico de ventas por producto.
Datos de ventas:

Producto	Ventas
Producto A	150
Producto B	200
Producto C	50
Producto A	100
Producto B	150
Producto C	100
Calcula:

El total de ventas.
El promedio de ventas por producto.
Genera un gráfico de barras de las ventas por producto.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Datos de ventas
data = {
    'Producto': ['Producto A', 'Producto B', 'Producto C', 'Producto A', 'Producto B', 'Producto C'],
    'Ventas': [150, 200, 50, 100, 150, 100]
}

# Crear un DataFrame
df = pd.DataFrame(data)

# 1. Calcular el total de ventas
total_ventas = df['Ventas'].sum()
print(f"Total de ventas: {total_ventas}")

# 2. Calcular el promedio de ventas por producto
promedio_ventas_por_producto = df.groupby('Producto')['Ventas'].mean()
print("Promedio de ventas por producto:")
print(promedio_ventas_por_producto)

# 3. Generar un gráfico de barras de las ventas por producto
ventas_por_producto = df.groupby('Producto')['Ventas'].sum()

plt.figure(figsize=(10, 6))
ventas_por_producto.plot(kind='bar')
plt.title('Ventas por Producto')
plt.xlabel('Producto')
plt.ylabel('Ventas Totales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
